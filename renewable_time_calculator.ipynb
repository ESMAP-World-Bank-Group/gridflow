```python
import math
```

```python
def estimate_and_confirm(num_subregions, points_per_subregion):
    """
    Calculate API time estimate and ask user confirmation to proceed.
    """
    # Constants are written by Renewables Ninja for the registered user account
    BURST_LIMIT = 6
    HOURLY_LIMIT = 50

    total_points = num_subregions * points_per_subregion
    total_queries = total_points * 2
    burst_minutes = math.ceil(total_queries / BURST_LIMIT)

    if total_queries > HOURLY_LIMIT:
        hourly_minutes = math.ceil(total_queries / HOURLY_LIMIT * 60)
    else:
        hourly_minutes = 0

    total_minutes = max(burst_minutes, hourly_minutes)


    hours = total_minutes // 60
    mins = total_minutes % 60
    time_str = f"{hours}h {mins}min"

    print(f"""
API Time Estimate:
- Total queries: {total_queries} ({total_points} points × 2)
- Estimated time: {time_str}
""")

    while True:
        response = input("Do you want to proceed? (yes/no): ").strip().lower()
        if response in ['yes', 'y']:
            print("Starting data collection now...")
            return
        elif response in ['no', 'n']:
            print("Data collection is cancelled")
            return False
        else:
            print("Please enter 'yes' or 'no'")
```

```python
result = estimate_and_confirm(17,5)
```
